### Section 1
Modify the functions in such a way that  
• **Medium Powered**: goes from 40, 50, 60  
• **Medium Fast**: goes from 50, 60, 70  
Determine for both cases (original sets and modified ones) the (approx.) values
for speed: 45 and 65.


```python
#Importing Necessary Packages
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
#Defining the Fuzzy Range from a speed of 30 to 90
x = np.arange(30, 80, 0.1)
#Defining the triangular membership functions
slow = fuzz.trimf(x, [30, 30, 50])
medium = fuzz.trimf(x, [30, 50, 70])
medium_fast = fuzz.trimf(x, [50, 60, 80])
full_speed = fuzz.trimf(x, [60, 80, 80])

# Original membership functions
plt.figure()
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed')
plt.plot(x, medium_fast, 'k', linewidth=1.5, label='Medium Fast')
plt.plot(x, medium, 'm', linewidth=1.5, label='Medium Powered')
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow')
plt.title('Penalty Kick Fuzzy Original')
plt.ylabel('Membership')
plt.xlabel("Speed (Miles Per Hour)")
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5),
ncol=1, fancybox=True, shadow=True)
plt.show()

# Modified membership functions
medium_m = fuzz.trimf(x, [40,50,60])
medium_fast_m = fuzz.trimf(x, [50, 60, 70])

plt.figure()
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed')
plt.plot(x, medium_fast_m, 'k', linewidth=1.5, label='Medium Fast')
plt.plot(x, medium_m, 'm', linewidth=1.5, label='Medium Powered')
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow')
plt.title('Penalty Kick Fuzzy Modified')
plt.ylabel('Membership')
plt.xlabel("Speed (Miles Per Hour)")
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5),
ncol=1, fancybox=True, shadow=True)
plt.show()
```

These are the results for the membership functions with the desired ranges. Now, to calculate our output values for the speeds 45 and 65 we do the following.

```python
# Define speeds
speeds = [45, 65]

# Original membership functions
print("Original Membership Functions")
slow_original = fuzz.interp_membership(x, slow, speeds[0])
medium_original = fuzz.interp_membership(x, medium, speeds[0])
medium_fast_original = fuzz.interp_membership(x, medium_fast, speeds[0])
fast_original = fuzz.interp_membership(x, full_speed, speeds[0])
print("Speed 45 mph")
print(f"Slow membership: {slow_original}")
print(f"Medium membership: {medium_original}")
print(f"Medium Fast membership: {medium_fast_original}")
print(f"Full speed  membership: {fast_original}")

slow_original = fuzz.interp_membership(x, slow, speeds[1])
medium_original = fuzz.interp_membership(x, medium, speeds[1])
medium_fast_original = fuzz.interp_membership(x, medium_fast, speeds[1])
fast_original = fuzz.interp_membership(x, full_speed, speeds[1])
print("Speed 65 mph")
print(f"Slow membership: {slow_original}")
print(f"Medium membership: {medium_original}")
print(f"Medium Fast membership: {medium_fast_original}")
print(f"Full speed  membership: {fast_original}")


# Modified functions
print()
print("Modified membership functions")
slow_modified = fuzz.interp_membership(x, slow, speeds[0])
medium_modified = fuzz.interp_membership(x, medium_m, speeds[0])
medium_fast_modified = fuzz.interp_membership(x, medium_fast_m, speeds[0])
fast_modified = fuzz.interp_membership(x, full_speed, speeds[0])
print("Speed 45 mph")
print(f"Slow membership: {slow_modified}")
print(f"Medium membership: {medium_modified}")
print(f"Medium Fast membership: {medium_fast_modified}")
print(f"Full speed  membership: {fast_modified}")

slow_modified = fuzz.interp_membership(x, slow, speeds[1])
medium_modified = fuzz.interp_membership(x, medium_m, speeds[1])
medium_fast_modified = fuzz.interp_membership(x, medium_fast_m, speeds[1])
fast_modified = fuzz.interp_membership(x, full_speed, speeds[1])
print("Speed 65 mph")
print(f"Slow membership: {slow_modified}")
print(f"Medium membership: {medium_modified}")
print(f"Medium Fast membership: {medium_fast_modified}")
print(f"Full speed  membership: {fast_modified}")
```

### Trapezoidal Membership Function

With the soccer example, you have to provide a range based on a specific class. In this
membership function, the membership reaches 100% from 0% in the center, and then
again drops to 0%. Instead of three points, as with the triangular membership function,
you have four points.  
This applies the soccer example to the trapezoidal membership function, whose classes
are defined as follows:  
• **Full speed** as [60, 80, 80, 90]  
• **Medium powered**as [30, 50, 50, 70]  
• **Slow** as [20, 30, 30, 50]  
• **Medium** fast as [50, 60, 60, 80]  

```python
#Defining the Fuzzy Range from a speed of 30 to 90
x = np.arange(30, 120, 0.1)
#Defining the trapezoidal membership functions
slow = fuzz.trapmf(x, [30, 40, 50, 60])
medium = fuzz.trapmf(x, [50, 60, 70, 80])
medium_fast = fuzz.trapmf(x, [70, 80, 90, 100])
full_speed = fuzz.trapmf(x, [90, 100, 110, 120])
#Plotting the Membership Functions Defined
plt.figure()
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed')
plt.plot(x, medium_fast, 'k', linewidth=1.5, label='Medium Fast')
plt.plot(x, medium, 'm', linewidth=1.5, label='Medium Powered')
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow')
plt.title('Penalty Kick Fuzzy')
plt.ylabel('Membership')
plt.xlabel("Speed (Miles Per Hour)")
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5),
ncol=1, fancybox=True, shadow=True)
```

### Gaussian Membership Function
When you apply the Gaussian membership function to the soccer example, you can see
that the values are represented much better and the interpolation is smooth. You define
the Gaussian membership of the classes as follows:   

• **Full speed** has a mean of 80 mph and a standard deviation of 4  
• **Medium powered** has a mean of 50 mph and a standard deviation of 4  
• **Slow** has a mean of 30 mph and a standard deviation of 4  
• **Medium** fast has a mean of 60 mph and a standard deviation of 4 

You can always play with the standard deviation. Here is the Python implementation of
the soccer example using the Gaussian membership function.

```python
# Importing Necessary Packages
# Defining the Fuzzy Range from a speed of 30 to 90
x = np.arange(30, 90, 0.1)
#Defining the gaussian membership functions
full_speed = fuzz.gaussmf(x, 80, 4)
medium_fast = fuzz.gaussmf(x, 60, 4)
medium = fuzz.gaussmf(x, 50, 4)
slow = fuzz.gaussmf(x, 30, 4)
#Plotting the Membership Functions Defined
plt.figure()
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed')
plt.plot(x, medium_fast, 'k', linewidth=1.5, label='Medium Fast')
plt.plot(x, medium, 'm', linewidth=1.5, label='Medium Powered')
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow')
plt.title('Penalty Kick Fuzzy Original')
plt.ylabel('Membership')
plt.xlabel("Speed (Miles Per Hour)")
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5),
ncol=1, fancybox=True, shadow=True)

#Plotting the Membership Functions Defined
medium_fast_m = fuzz.gaussmf(x, 60, 10)
medium_m = fuzz.gaussmf(x, 50, 12)

plt.figure()
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed')
plt.plot(x, medium_fast_m, 'k', linewidth=1.5, label='Medium Fast')
plt.plot(x, medium_m, 'm', linewidth=1.5, label='Medium Powered')
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow')
plt.title('Penalty Kick Fuzzy Modified')
plt.ylabel('Membership')
plt.xlabel("Speed (Miles Per Hour)")
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5),
ncol=1, fancybox=True, shadow=True)
```

### Generalized Bell Membership Function
If you solve the soccer example with the generalized bell function, you’ll get the
following membership functions:  

• **Full speed** has a center at 80 mph while the width and slope are 8 and 4,
respectively  
• **Medium powered** has a center at 50 mph while the width and slope are 8 and 4,
respectively  
• **Slow** has a center at 30 mph while the width and slope are 8 and 4, respectively  
• **Medium** fast has a center at 60 mph while the width and slope are 8 and 4,
respectively 

Modify such that the center of the bell for Full Speed goes at 85, and separate
the centers of medium fast and medium powered is of 20 units.


```python
#Defining the Fuzzy Range from a speed of 30 to 90
x = np.arange(30, 90, 0.1)
#Defining the generalized bell membership functions
full_speed = fuzz.gbellmf(x, 8,4,85)
medium_fast = fuzz.gbellmf(x, 8,4,60)
medium = fuzz.gbellmf(x, 8,4,40)
slow = fuzz.gbellmf(x, 8,4,30)
#Plotting the Membership Functions Defined
plt.figure()
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed')
plt.plot(x, medium_fast, 'k', linewidth=1.5, label='Medium Fast')
plt.plot(x, medium, 'm', linewidth=1.5, label='Medium Powered')
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow')
plt.title('Penalty Kick Fuzzy')
plt.ylabel('Membership')
plt.xlabel("Speed (Miles Per Hour)")
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5),
ncol=1, fancybox=True, shadow=True)
```

### Sigmoidal Membership Function
When you come to a specific instance wherein you must take care of very high values or
very low values, you use the sigmoidal membership function as the target. You must
provide two points, and the most important point is the c point (crossover point), which
represents the center.  

Therefore, for the soccer example, the classes will be redefined using the sigmoidal
membership function as follows:  

• **Full speed** has a crossover at 80 mph and a slope of 2   
• **Medium powered** has a crossover at 50 mph and a slope of 2  
• **Slow** has a crossover at 30 mph and a slope of 2  
• **Medium fast** has a crossover at 60 mph and a slope of 2  

```python
# Defining the Fuzzy Range from a speed of 30 to 90
x = np.arange(30, 90, 0.1)
#Defining the sigmoidal membership functions
full_speed = fuzz.sigmf(x, 80,2)
medium_fast = fuzz.sigmf(x, 60,2)
medium = fuzz.sigmf(x, 50,2)
slow = fuzz.sigmf(x, 30,2)
#Plotting the Membership Functions Defined
plt.figure()
plt.plot(x, full_speed, 'b', linewidth=1.5, label='Full Speed')
plt.plot(x, medium_fast, 'k', linewidth=1.5, label='Medium Fast')
plt.plot(x, medium, 'm', linewidth=1.5, label='Medium Powered')
plt.plot(x, slow, 'r', linewidth=1.5, label='Slow')
plt.title('Penalty Kick Fuzzy')
plt.ylabel('Membership')
plt.xlabel("Speed (Miles Per Hour)")
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5),
ncol=1, fancybox=True, shadow=True)
```
